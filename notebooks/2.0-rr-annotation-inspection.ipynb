## Number of segments per turn

```python
train_dataframe['num_segments'] = train_dataframe['segments'].apply(len)
```

```python
train_dataframe['num_segments'].describe()
```

Most turns have approximately 2-3 segments. It's surprising there's an outlier turn with several segments. What's up with that?

```python
train_dataframe['num_segments'].hist()
```

### Long segment turns review.

```python
train_dataframe[train_dataframe['num_segments'] > 8][['conversation_id', 'turn_index', 'message']]
```

These utterances don't seem out of place or noisy, but they're incredibly rambly

```python
train_dataframe.groupby(["conversation_id"])[["turn_index", "message"]].agg(list)
```

```python
train_dataframe['segment_to_turn_idx'] = train_dataframe.apply(lambda row: [row['turn_index'] for _ in row['segments']], axis=1)
```

```python
train_dataframe.groupby(["conversation_id"])['segment_to_turn_idx'].agg(list).apply(lambda turns: [idx for turn in turns for idx in turn])
```

```python
grouped_conversation_data = train_dataframe.groupby(["conversation_id"])['message','segment_to_turn_idx'].agg(list)
```

```python
grouped_conversation_data.apply(lambda turns: [idx for turn in turns for idx in turn]).apply(pd.Series.explode)
```

```python
grouped_conversation_data['flattened_segment_turn_list'] = grouped_conversation_data['segment_to_turn_idx'].agg(list).apply(lambda turns: [idx for turn in turns for idx in turn]).apply(pd.Series.explode)
```
